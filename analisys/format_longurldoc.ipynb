```python
#%pip install pandas
#%pip install matplotlib

import json
import pandas as pd
import matplotlib.pyplot as plt

question_types = {}
evidence_sources = {}
answer_formats = {}
tasks_tag = {}
count = 0
with open("LongDocURL_public.jsonl", "r", encoding="utf-8", errors="replace") as file:
    for line in file:
        count += 1
        
        try:
            data = json.loads(line)
            #print(data)

            qtype = data['question_type']
            dict_question = question_types.keys()
            if qtype in dict_question:
                question_types[qtype] = question_types[qtype] + 1
            else:
                question_types[qtype] = 1

            evidence_src = data['evidence_sources']
            #print(f'evidence_src: {evidence_src}')
            for ev in evidence_src:
                if ev is not None or ev != " ":
                    #print(f"ev: {ev}")
                    if ev in evidence_sources:
                        evidence_sources[ev] = evidence_sources[ev] + 1
                    else:
                        evidence_sources[ev] = 1

            answr_format = data['answer_format']
            dict_answer_formats = answer_formats.keys()
            if answr_format in dict_answer_formats:
                answer_formats[answr_format] = answer_formats[answr_format] + 1
            else:
                answer_formats[answr_format] = 1

            task_tag = data['task_tag']
            dict_task_tag = tasks_tag.keys()
            if task_tag in dict_task_tag:
                tasks_tag[task_tag] = tasks_tag[task_tag] + 1
            else:
                tasks_tag[task_tag] = 1
        except:
            print("linha com erro de conversão ou leitura: ", count)
            
            
        
print(f"question_types: {type(question_types)} - {question_types}")
print(f"evidence_sources: {type(evidence_sources)} - {evidence_sources}")
print(f"answer_formats: {type(answer_formats)} - {answer_formats}")
print(f"tasks_tag: {type(tasks_tag)} - {tasks_tag}")
print(count)
```

```python
#QUESTION TYPES
df = pd.DataFrame(list(question_types.items()), columns=['Question Type', 'Quantidade'])

fig, ax = plt.subplots(figsize=(6, 4))
ax.axis('off')

tabela = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

tabela.auto_set_font_size(False)
tabela.set_fontsize(10)
tabela.scale(1.2, 1.2)

plt.show()
```

```python
#EVIDENCE SOURCES
df = pd.DataFrame(list(evidence_sources.items()), columns=['Evidence Source', 'Quantidade'])

fig, ax = plt.subplots(figsize=(6, 4))
ax.axis('off')

tabela = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

tabela.auto_set_font_size(False)
tabela.set_fontsize(10)
tabela.scale(1.2, 1.2)

plt.show()
```

```python
#ANSWER FORMATS
df = pd.DataFrame(list(answer_formats.items()), columns=['Answer formats', 'Quantidade'])

fig, ax = plt.subplots(figsize=(6, 4))
ax.axis('off')

tabela = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

tabela.auto_set_font_size(False)
tabela.set_fontsize(10)
tabela.scale(1.2, 1.2)

plt.show()
```
